```python
#find aggregated mean of the “order_number” column grouped by “department_id”
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

ii)The mean value is different for each department, for instance 'dept_id 4' had a mean value of 18.83 in the subset but has a   value of 17.81 in the dataframe, 'dept_id 17' has a value of 11.29 in the subset but has a value of 15.69 in dataframe

```python
#create new column containing the maximum frequency of the “order_number” 
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(10)
```

```python
ords_prods_merge.head(100)
```

Create column containing a flag designating whether a customer is “loyal”, "Regular" or "New"

```python
#Create 'Loyal customer'column
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#Create 'Regular customer'column
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#Create 'New customer'column
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#print frequency of “loyalty_flag” 
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(40)
```

```python
# Aggregating mean price by customer loyalty type
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

Spending Habits:
New customers spent on average 13.2
Regular customers spent on average 12.5
Loyal customers spent 10.4

Create column containing types of spenders flag

```python
#create 'mean_prices' column
ords_prods_merge['mean_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(10)
```

Creating spending_flag

```python
#Create 'Low spender'flag
ords_prods_merge.loc[ords_prods_merge['mean_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#Create 'High spender'flag
ords_prods_merge.loc[ords_prods_merge['mean_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#print frequency of “spending_flag” 
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

Create column containing order frequency flag

```python
#create 'median_days_since_prior_order' column
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(10)
```

Creating order frequency flag

```python
#Create 'Non-frequent customer'flag
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#Create 'Regular customer'flag
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
#Create 'Frequent customer'flag
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <=10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#print frequency of “order frequency flag” 
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
