Import libraries

```python
#import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

Import Data

```python
#Create path
path = r'C:\Users\mngit\OneDrive\Data Analytics\Project Excercises\Achievement 4\Instacart Basket Analysis' 
```

```python
# import orders_products_customers
orders_products_customers = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_customers.pkl'))
```

Creating sample subset

```python
#set the seed 
np.random.seed(4)
```

```python
#store result of test in new list
dev = np.random.rand(len(orders_products_customers)) <= 0.7
```

```python
#store 70% of the sample in the dataframe big
big = orders_products_customers[dev]
```

```python
#store 30% of the sample in the dataframe small
small = orders_products_customers[~dev]
```

_Check sample_

```python
len(orders_products_customers)
```

```python
len(big) + len(small)
```

```python
#reduce sample by selection of only required columns
df_2 = small[['orders_day_of_week','prices']]
```

Creating Visualizations

```python
#create line chart using subset- 'orders_day_of_week' by'prices'
line_1 = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
#Export line chart 'orders_day_of_week' by'prices'
line_1.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_prices_orders_dow.png'))
```

```python
#create bar chart of 'orders_day_of_week'
bar_1 = orders_products_customers['orders_day_of_week'].value_counts().plot.bar(color =['magenta', 'purple', 'blue', 'cyan', 'green', 'yellow', 'orange'])
```

```python
#Export bar chart - 'orders_day_of_week'
bar_1.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
#create histogram of 'prices' with 25 bars
hist_1 = orders_products_customers['prices'].plot.hist(bins = 25)
```

```python
#Export histogram - 'prices'
hist_1.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_orders_prices.png'))
```

```python
#create scatterplot of 'prices' 

plt.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scarter_plot_prices1.png'))
plt.show()
sns.scatterplot(x = 'prices', y = 'prices',data = orders_products_customers)
```

```python
#Export scatterplot - 'prices' 
plt.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scarter_plot_prices1.png'))
plt.show()
```

```python
#create histogram of 'order_hour_of_day' with 24 bars
hist_2=orders_products_customers['order_hour_of_day'].plot.hist(bins = 24)
```

Above Histogram demonstares that the busiest hour of the day is 10:00am as indicated by the tallest bar. The busiest period is roughly between 10:00 and 17:00

```python
#create bar chart of 'loyalty_flag'
bar_2 = orders_products_customers['loyalty_flag'].value_counts().plot.bar(color =['purple', 'blue', 'cyan'])
```

```python
#Export bar chart -'loyalty_flag'
bar_2.figure.savefig(os.path.join(path, '04 Analysis','Visualizations','bar_loyalty.png'))
```
